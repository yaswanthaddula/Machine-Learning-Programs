```python
import pandas as pd
```

```python
df = pd.read_csv("/content/dataset.csv")
df
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
n = X.shape[1]
```

```python
S = ['∅'] * n
G = [['?'] * n]
```

```python

for i in range(len(X)):
    if y[i] == 'Yes':
        for j in range(n):
            if S[j] == '∅':
                S[j] = X[i][j]
            elif S[j] != X[i][j]:
                S[j] = '?'
        G = [g for g in G if all(g[k] == '?' or g[k] == S[k] for k in range(n))]
    else:
        new_G = []
        for g in G:
            for j in range(n):
                if g[j] == '?' and S[j] != X[i][j]:
                    h = g.copy()
                    h[j] = S[j]
                    new_G.append(h)
        G = new_G
```

```python
print("Final S:", S)
print("Final G:", G)
```
