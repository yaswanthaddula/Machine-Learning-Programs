```python
# Implement House price Prediction using appropriate machine learning algorithm
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
```

```python
house=pd.read_csv('/content/data.csv')
house
```

```python
df=house.drop(['street','city','statezip','country','date'],axis=1)
```

```python
y=df['price']
```

```python
x=df.drop(['price'],axis=1)
```

```python
x_test,x_train,y_test,y_train=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
linear1=LinearRegression()
linear1.fit(x_train,y_train)
```

```python
y_pred=linear1.predict(x_test)
mse1=mean_squared_error(y_test,y_pred)
print(mse1)
```
