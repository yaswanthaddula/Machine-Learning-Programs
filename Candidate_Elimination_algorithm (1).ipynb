```python
# Candidate Elimination - Simple Version

data = [
    ['Sunny', 'Warm', 'Normal', 'Strong', 'Warm', 'Same', 'Yes'],
    ['Sunny', 'Warm', 'High', 'Strong', 'Warm', 'Same', 'Yes'],
    ['Rainy', 'Cold', 'High', 'Strong', 'Warm', 'Change', 'No'],
    ['Sunny', 'Warm', 'High', 'Strong', 'Cool', 'Change', 'Yes']
]

X = [row[:-1] for row in data]
y = [row[-1] for row in data]

n = len(X[0])

S = ['0'] * n
G = [['?'] * n]

print("Initial S:", S)
print("Initial G:", G)


for i in range(len(X)):

    if y[i] == "Yes":

        for j in range(n):
            if S[j] == '0':
                S[j] = X[i][j]
            elif S[j] != X[i][j]:
                S[j] = '?'

        G = [g for g in G if all(g[k] == '?' or g[k] == X[i][k] for k in range(n))]

    else:

        new_G = []
        for g in G:
            for j in range(n):
                if g[j] == '?':
                    if S[j] != X[i][j]:
                        new_hyp = g.copy()
                        new_hyp[j] = S[j]
                        new_G.append(new_hyp)
        G = new_G

    print("\nAfter Example", i+1)
    print("S:", S)
    print("G:", G)

print("\nFinal S:", S)
print("Final G:", G)
```
