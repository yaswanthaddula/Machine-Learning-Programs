```python
# Implement the Car Price Prediction Model using Python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
car=pd.read_csv('/content/car_price_prediction_.csv',delimiter=',')
print(car)
```

```python
car.isnull().sum()
```

```python
df=car.drop(['Car ID','Model'],axis=1)
```

```python
y = df['Price']
```

```python
x = df.drop('Price', axis=1)
```

```python
x = pd.get_dummies(x, drop_first=True)
```

```python
x_test,x_train,y_test,y_train=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
linear=LinearRegression()
linear.fit(x_train, y_train)
```

```python
y_pred=linear.predict(x_test)
```

```python
new_car = {
    'Brand': 'Tesla',
    'Year': 2020,
    'Engine Size': 2.5,
    'Fuel Type': 'Petrol',
    'Transmission': 'Automatic',
    'Mileage': 30000,
    'Condition': 'New'
}
```

```python
new_df = pd.DataFrame([new_car])
```

```python
new_df
```

```python
new_df_encode=pd.get_dummies(new_df,drop_first=True)
new_df_encode = new_df_encode.reindex(columns=x_train.columns,fill_value=0)
```

```python
predicted_price = linear.predict(new_df_encode)
print("Predicted Car Price:", predicted_price)
```
