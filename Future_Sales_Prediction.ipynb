```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report
```

```python
df=pd.read_csv('/content/Ecommerce_Sales_Prediction_Dataset.csv')
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
```

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
```

```python
df
```

```python
df['Product_Category'].unique()
#df['Customer_Segment'].unique()
```

```python
df['Customer_Segment'] = df['Customer_Segment'].map({'Occasional': 0,'Premium': 1,'Regular': 2})
df['Product_Category'] = df['Product_Category'].map({'Sports': 1,'Toys': 2,'Home Decor': 3,'Fashion':4,'Electronics':5})
```

```python
x= df.drop('Units_Sold',axis=1)
x=df.drop('Date',axis=1)
x
```

```python
y=df['Units_Sold']
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
accu=accuracy_score(y_test,y_pred)
print("Accuracy_Score:",accu)
```

```python
y_pred=model.predict(x_test)
cr=classification_report(y_test,y_pred)
print("Classification_Report:\n",cr)
```
