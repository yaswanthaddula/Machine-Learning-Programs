```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('/content/Credit Score Classification Dataset.csv')
df
```

```python
df['Education'].unique()
```

```python
df.isnull().sum()
```

```python
df['Gender']=df['Gender'].map({'Male':0,'Female':1})
```

```python
df['Marital Status']=df['Marital Status'].map({'Married':1,'Single':0})
```

```python
df['Credit Score']=df['Credit Score'].map({'Low':0,'Average':1,'High':2})
```

```python
df['Home Ownership']=df['Home Ownership'].map({'Rented':3,'Owned':4})
```

```python
df["Education"] = df["Education"].map({"High School Diploma": 0,"Associate's Degree": 1,"Bachelor's Degree": 2,"Master's Degree": 3,"Doctorate": 4
})
```

```python
df
```

```python
x=df.iloc[:,0:7]
x
```

```python
y=df.iloc[:,7:]
y
```

```python
x_test,x_train,y_test,y_train=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy_score:",accuracy)
```

```python
sample=[[40,0,125000,0,0,0,4]]
prediction=model.predict(sample)
print("Sample Prediction:",prediction)
```
