```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df=pd.read_csv('/content/Mobile Price Prediction Datatset.csv')
df
```

```python
df.isnull().sum()
```

```python
df.drop(['Brand me'],axis=1,inplace=True)
```

```python
df.fillna(df.mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop(['Unnamed: 0'],inplace=True,axis=1)
```

```python
df
```

```python
x=df.iloc[:,0:7]
x
```

```python
y=df.iloc[:,7:]
y
```

```python
x_test,x_train,y_test,y_train=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
meansquare=mean_squared_error(y_test,y_pred)
print("Mean_Squarred_Error:",meansquare)
r2=r2_score(y_test,y_pred)
print("R2 Square:",r2)
```

```python
new=[[3.8,6.0,32.0,4.54,48,12.000000,2800]]
prediction=model.predict(new)
print("Sample Prediction:",prediction)
```
