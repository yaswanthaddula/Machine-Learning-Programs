```python
# Implement the Car Price Prediction Model using Python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
car=pd.read_csv('/content/car_price_prediction_.csv',delimiter=',')
print(car)
```

```python
car.isnull().sum()
```

```python
car.duplicated().sum()
```

```python
car1=car.drop(['Car ID','Model'],axis=1)
print(car1)
```

```python
x=car1.iloc[:,0:7]
print(x)
```

```python
y=car1.iloc[:,7:]
print(y)
```

```python
x_test,x_train,y_test,y_train=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train_encoded = pd.get_dummies(x_train, drop_first=True)
x_test_encoded = pd.get_dummies(x_test, drop_first=True)
```

```python
train_columns = x_train_encoded.columns
x_test_encoded = x_test_encoded.reindex(columns=train_columns,fill_value=0)
```

```python
linear=LinearRegression()
linear.fit(x_train_encoded,y_train)
```

```python
y_pred=linear.predict(x_test_encoded)
mse=mean_squared_error(y_test,y_pred)
print(mse)
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
print(score)
```
